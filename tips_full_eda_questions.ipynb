# üìò Exploratory Data Analysis on Seaborn Tips Dataset
This notebook contains problem statements for Univariate, Bivariate, and Multivariate analysis. Write your own code below each question to explore the dataset.

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np
import seaborn as sns
tips = sns.load_dataset('tips')
tips.head()
```

# üîç Univariate Analysis

### Q1. What is the distribution of the `total_bill`?

```python
plt.hist(tips['total_bill'],edgecolor='black')
plt.title('Distribution of total_bills')
plt.show()
```

### Q2. What is the distribution of `tip` amounts?

```python
plt.hist(tips['tip'],edgecolor='black',bins=20)
plt.title('Distribution of tip')
plt.gca().set_facecolor('#dfea8c')
plt.gcf().set_facecolor('#7e6e6b')
plt.tight_layout()
plt.show()
```

### Q3. How is the group `size` distributed?

```python
sns.countplot(tips,y=tips['size'],edgecolor='black',color='grey')
plt.gca().set_facecolor('#c697c6')
plt.gcf().set_facecolor('#e257e4')
plt.show()
```

### Q4. How many males vs females visited?

```python
colors =['#b7c4ee','#e969f1']
sns.countplot(tips['sex'].value_counts(),palette=colors,edgecolor='black')
plt.gca().set_facecolor('#f6f1ab')
```

### Q5. What is the proportion of smokers vs non-smokers?

```python
proportions = tips['smoker'].value_counts(normalize=True)
labels = proportions.index
sizes = proportions.values
plt.pie(sizes, labels=labels, autopct='%.1f%%', startangle=180, colors=['#66b3ff','#ff9999'],explode=[0.1,0])
plt.title('Proportion of Smokers vs Non-Smokers')
plt.show()
```

### Q6. What are the counts of visits per day?

```python
sns.countplot(data=tips, x='day', order=['Thur', 'Fri', 'Sat', 'Sun'])
plt.title('Number of Visits per Day')
plt.xlabel('Day')
plt.ylabel('Count')
plt.show()
```

### Q7. How often do people dine at Lunch vs Dinner?

# üîç Bivariate Analysis

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np
import seaborn as sns
tips = sns.load_dataset('tips')
tips.head()
```

### Q1. Is there a correlation between `total_bill` and `tip`?

```python
sns.scatterplot(data=tips, x='total_bill', y='tip', color='skyblue')
plt.title('Correlation between Total Bill and Tip')
plt.xlabel('Total Bill ')
plt.ylabel('Tip ')
plt.show()
```

### Q2. Do men tip more than women?

```python
y=tips.groupby('sex')['tip'].sum()
colors =['#b7c4ee','#e969f1']
plt.pie(y,labels=y.index,autopct='%2f')
plt.show()
```

### Q3. How does `tip` vary by day?

```python
sns.boxplot(data=tips, x='day', y='tip', color='purple')
plt.title('Tip by Day')
plt.xlabel('Day of the Week')
plt.ylabel('Tip ')
plt.show()
```

### Q4. Compare average `total_bill` for smokers vs non-smokers.

```python
sns.boxplot(x="smoker", y="total_bill", data=tips)
plt.title("Total Bill Distribution: Smokers vs Non-Smokers")
plt.ylabel("Total Bill ($)")
plt.xlabel("Smoker")
plt.show()
```

### Q5. How does `tip` vary across different meal times?

```python
import matplotlib.pyplot as plt
sns.boxplot(x="time", y="tip", data=tips,color='lightcoral')
plt.title("Tip Distribution by Meal Time")
plt.xlabel("Meal Time")
plt.ylabel("Tip ($)")
plt.show()
```

### Q6. Relationship between `size` and `tip`?

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(x="size", y="tip", data=tips)
plt.gca().set_facecolor('#f9f9f9')
plt.gcf().set_edgecolor('#eaeaf2')
plt.title("Tip Amount vs Party Size")
plt.xlabel("Party Size")
plt.ylabel("Tip Amount ($)")
plt.show()
```

### Q7. Total bill trend by day of week

```python
plt.figure(figsize=(8, 5))
sns.barplot(x="day", y="total_bill", data=tips, palette="pastel")
plt.title("Average Total Bill by Day of the Week", color='darkblue')
plt.xlabel("Day", color='black')
plt.ylabel("Average Total Bill", color='black')
plt.show()
```

### Q8. Gender distribution over days

```python
plt.figure(figsize=(8, 5))
sns.countplot(x="day",hue="sex", data=tips, palette="pastel")
plt.title("Gender Distribution Over Days", color='darkblue')
plt.xlabel("Day of the Week", color='black')
plt.ylabel("Number of Customers", color='black')
```

### Q9. Average tips for different group sizes

### Q10. Do males smoke more than females?

# üîç Multivariate Analysis

### Q1. How do `total_bill` and `tip` relate across gender and smoker status?

```python
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='sex', style='smoker')
plt.suptitle('Total Bill and Tip Relation Across Gender and Smoker Status')
plt.show()
```

### Q2. What is the correlation between all numeric variables?

```python
tips.select_dtypes(include='number').corr()
sns.heatmap(tips.select_dtypes(include='number').corr(),annot=True)
plt.title('The heatmap of numeric values')
plt.gca().set_facecolor('#f6f1ab')
plt.gcf().set_facecolor('#8b8a7c')
plt.show()
```

### Q3. How does tip vary by gender, day, and smoker status?

```python
fig, axes = plt.subplots(1, 3, figsize=(18, 6))
sns.boxplot(data=tips, x='sex', y='tip', ax=axes[0])
axes[0].set_title('Tip by Gender')
plt.gca().set_facecolor('#f6f1')
plt.gcf().set_facecolor('#8b8a7c')

sns.boxplot(data=tips, x='day', y='tip', ax=axes[1])
axes[1].set_title('Tip by Day')
plt.gca().set_facecolor('#f6f1ab')
plt.gcf().set_facecolor('#8b8a7c')

sns.boxplot(data=tips, x='smoker', y='tip', ax=axes[2])
axes[2].set_title('Tip by Smoker Status')

plt.show()
```

### Q4. Do men tip more than women on weekends?

```python
fig, axes = plt.subplots(1, 3, figsize=(18, 6))
sns.boxplot(data=tips, x='sex', y='tip', ax=axes[0],color='green')
axes[0].set_title('Tip by Gender')
plt.gca().set_facecolor('#f6f1ab')
plt.gcf().set_facecolor('#8b8a7c')

sns.boxplot(data=tips, x='day', y='tip', ax=axes[1],color='yellow')
axes[1].set_title('Tip by Day')
plt.gca().set_facecolor('#f6f1ab')
plt.gcf().set_facecolor('#8b8a7c')

sns.boxplot(data=tips, x='smoker', y='tip', ax=axes[2],color='lightgreen')
axes[2].set_title('Tip by Smoker Status')
plt.show()
```

### Q5. Heatmap of average tip by day and time

```python
fig, axes = plt.subplots(1, 2, figsize=(18, 6))
sns.barplot(data=tips, x='sex', y='tip', ax=axes[0],color='skyblue')
axes[0].set_title("Average Tip by Gender", fontsize=13, weight='bold')
axes[0].set_xlabel("Gender")
axes[0].set_ylabel("Average Tip ")
plt.gca().set_facecolor('#f6f1ab')
plt.gcf().set_facecolor('#8b8a7c')
sns.boxplot(data=tips, x='sex', y='tip', palette='Pastel1', ax=axes[1])
axes[1].set_title("Tip Distribution by Gender on Weekends", fontsize=13, weight='bold')
axes[1].set_xlabel("Gender")
axes[1].set_ylabel("Tip Amount ")
plt.suptitle("Do Men Tip More Than Women on Weekends", fontsize=16, weight='bold')
plt.show()
```
